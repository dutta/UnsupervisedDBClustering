```python
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans, SpectralClustering, AgglomerativeClustering
import seaborn as sns
# Create the default pairplot

# THIS IS THE FILE THAT WAS MADE FROM THE FEATURE FILE !!CHANGE THE PATH!!#
df_safeties = pd.read_csv("/users/rishav/documents/Big-Data-Bowl/Data/corners_features_with_ratios.csv").dropna()
print(df_safeties.keys())

# LOOK AT JUST THE FEATURES AND IGNORE THE EXTRA FEATURES THAT ARENT INVOLVED IN THE CLUSTERING#
df_safeties_to_cluster = df_safeties[['VAR_X',
                                      'VAR_Y',
                                      'SPEED_VAR',
                                      'OFF_VAR',
                                      'OFF_MEAN',
                                      'DEF_VAR',
                                      'DEF_MEAN',
                                      'OFF_DIR_VAR',
                                      'OFF_DIR_MEAN',
                                      '0-DIS_OFF',
                                      '2-DIS_OFF',
                                      '4-DIS_OFF', 
                                      'RAT-type2-MEAN', 
                                      'RAT-type2-MID',
                                      'RAT-type2-PASS',
                                      'RAT-type2-SNAP', 
                                      'RAT-type2-VAR'
                                         ]] 


#GET JUST THE VALUES IN A MATRIX#
mat = df_safeties_to_cluster.values

#CREATE THE GMM WITH 2 CLUSTERS#
gmm = GaussianMixture(n_components=2)

#CREATE THE HC MODEL#
clustering = AgglomerativeClustering().fit(mat)

#FIT THE DATA TO THE GMM#
preds = gmm.fit(mat)

#GET THE PREDICTION PROBABILITIES#
label1 = gmm.predict_proba(mat)

#FUNCTION TO GET THE CLUSTER THAT WAS GREATER#
def get_cluster(row):
    if(row['cluster 0'] > row['cluster 1']):
        return 'cluster 0'
    return 'cluster 1'

#APPEND THE CLUSTER VALUES TO THE ORIGINAL DATAFRAME#
results = pd.DataFrame(data=label1, columns=['cluster 0','cluster 1'])
higherarch = pd.DataFrame(data = clustering.labels_, columns = ['Hierarchical Clustering'])
predy = results.apply(get_cluster,axis=1)
new_df = df_safeties.join(pd.DataFrame({'cluster 0' : results['cluster 0'].round(3),'cluster 1' : results['cluster 1'].round(3), 'pred' : predy}))
new_df = new_df.join(higherarch)
print(new_df)

#POST THE DATA TO THE FILE !!CHANGE THE PATH!!#
new_df.to_csv("/users/rishav/documents/Big-Data-Bowl/Data/corners_with_clusters_ratios_1.csv")
print('Done')
```
